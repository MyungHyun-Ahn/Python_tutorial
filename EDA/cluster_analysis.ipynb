```python
# 클러스터 분석 Cluster : 무리 비슷한 것끼리 묶어서 분석
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
sur_df = pd.read_csv('data/young_survey.csv')
sur_df.head()
```

```python
interest_df = sur_df.loc[:, 'History':'Pets']
interest_df.head()
```

```python
inter_corr = interest_df.corr()
```

```python
inter_corr
```

```python
inter_corr['History'].sort_values(ascending=False)
```

```python
sns.clustermap(data=inter_corr)
```
