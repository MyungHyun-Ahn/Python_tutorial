```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
body_df = pd.read_csv('data/body.csv')
body_df.head()
```

```python
body_df.plot(kind='scatter', x='Height', y='Weight')
```

```python
sns.lmplot(data=body_df, x='Height', y='Weight')
# 가운데 선 회귀선 Regression line 흩어져 있는 점들을 선으로 표현
# 몸무게와 키와 연관성이 크지 않아 정확도 낮음
```
